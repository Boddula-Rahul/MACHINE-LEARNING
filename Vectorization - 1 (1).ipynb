```python
import numpy as np
```

```python
import pandas as pd
```

```python
sent = ["Brown Dog Chasing White rabbit",
       "White Rabbit run into Green gate",
       "Dog jumped over Green ate to catch rabbit",
       "Brown og lost rabbit in green grass"]
```

```python
sent
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cvec = CountVectorizer()
```

```python
cvec
```

```python
cvec.fit(sent)
```

```python
cvec.vocabulary_
```

```python
cvec.get_feature_names_out()
```

```python
cvec.transform(sent)
```

```python
ary = cvec.transform(sent).toarray()
```

```python
ary
```

```python
sent
```

```python
cvec.get_feature_names_out()
```

```python
pd.DataFrame(ary,columns = cvec.get_feature_names_out())
```
