```python
for i in range (61) :
    sk = df[i].skew().round()
    if sk == 0 or sk == -0 :
        m = df[i].mean()
        s = df[i].std()
        lower = m - 3*s
        upper = m + 3*s
        df[i] = df[i].map(lambda x : lower if x < lower else upper if x > upper else x )

else :
    q1 = df[i].quantile(0.25)
    q3 = df[i].quantile(0.75)
    iqr = q3 - q1
    low = q1 - (1.5*iqr)
    high = q3 + (1.5*iqr)
    df[i] = df[i].map(lambda x : low if x < low else high if x > high else x )
    
        
```
